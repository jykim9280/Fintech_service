```python
import requests
import pandas as pd
import os
from dotenv import load_dotenv
load_dotenv()
from datetime import datetime, date
import dbio
```

```python
url = "https://openapi.naver.com/v1/search/news.json"
payload = {'query' : '핀테크', 'display' : 100, 'start' : 1, 'sort' : 'sim'}
headers = {'X-Naver-Client-Id' : os.getenv('client_id'), 'X-Naver-Client-Secret' :os.getenv('client_secret')}
r = requests.get(url, params = payload, headers = headers)
print(r.url)
if(r.status_code ==200):
    data = r.json()
else:
    print("Error Code:" + r.status_code)
```

```python
data
```

```python
df = pd.json_normalize(data['items'])
df
```

```python
print(data['items'][0]['pubDate'])

today = date.today()
formatted_date = today.strftime("%d %b %y")
print(formatted_date)

pub_date_str = data['items'][0]['pubDate']
pub_date = datetime.strptime(pub_date_str, '%a, %d %b %Y %H:%M:%S %z')
pub_date_formatted = pub_date.strftime("%d %b %y")
print(pub_date_formatted)
```

```python
formatted_date in pub_date_formatted
```

```python
def text_clean(x):
    x = x.replace("&quot;", "").replace("</b>", "").replace("<b>", "")
    return x
```

```python
result = {}

for item in data['items']:
    if formatted_date in pub_date_formatted:
        for key in item.keys():
            result.setdefault(key, []).append(text_clean(item[key]))
    else:
        break

df = pd.DataFrame(result)
df
```

```python
dbio.news_to_db("fintech_news_test", str(today), df)
```
