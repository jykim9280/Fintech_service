```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(data, 'xml')

# 예시: item 태그 내의 모든 자식 태그 추출
items = soup.find_all('item')

# 추출한 데이터를 저장할 리스트 초기화
data_list = []

for item in items:
    data_dict = {}
    for child in item.find_all():  # 모든 자식 태그 반복
        data_dict[child.name] = child.text
    data_list.append(data_dict)

data_list
```

```python
value_list = [list(item.values()) for item in data_list]
columns = list(data_list[0].keys())
for values in value_list:
    df = pd.DataFrame(value_list, columns=columns)
df
```
