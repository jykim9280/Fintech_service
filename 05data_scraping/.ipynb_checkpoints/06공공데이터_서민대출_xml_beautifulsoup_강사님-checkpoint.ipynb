```python
import requests
import os
from dotenv import load_dotenv
import pandas as pd
import time
from bs4 import BeautifulSoup
load_dotenv()
```

# API 응답메세지 명세를 참고해 한땀한땀 만들기

```python
final_result = []
page = 1
while True:
    url = "http://apis.data.go.kr/1160100/service/GetSmallLoanFinanceInstituteInfoService/getOrdinaryFinanceInfo"
    payload = dict(serviceKey=os.getenv('servicekey'), pageNo=page, numOfRows=1000)
    r = requests.get(url, params=payload)
#     print(r.status_code)
    data = r.text
    soup = BeautifulSoup(data, 'xml')
    total_pages = int(soup.find('totalCount').string) // int(soup.find('numOfRows').string) + 1
    print(f"{page}/{total_pages} 수집중", end="\r")
    basYm = [item.string for item in soup.find_all('basYm')]
    snq = [item.string for item in soup.find_all('snq')]
    finPrdNm = [item.string for item in soup.find_all('finPrdNm')]
    lnLmt = [item.string for item in soup.find_all('lnLmt')]
    irtCtg = [item.string for item in soup.find_all('irtCtg')]
    irt = [item.string for item in soup.find_all('irt')]
    maxTotLnTrm = [item.string for item in soup.find_all('maxTotLnTrm')]
    maxDfrmTrm = [item.string for item in soup.find_all('maxDfrmTrm')]
    maxRdptTrm = [item.string for item in soup.find_all('maxRdptTrm')]
    rdptMthd = [item.string for item in soup.find_all('rdptMthd')]
    usge = [item.string for item in soup.find_all('usge')]
    trgt = [item.string for item in soup.find_all('trgt')]
    instCtg = [item.string for item in soup.find_all('instCtg')]
    ofrInstNm = [item.string for item in soup.find_all('ofrInstNm')]
    rsdAreaPamtEqltIstm = [item.string for item in soup.find_all('rsdAreaPamtEqltIstm')]
    suprTgtDtlCond = [item.string for item in soup.find_all('suprTgtDtlCond')]
    age = [item.string for item in soup.find_all('age')]
    incm = [item.string for item in soup.find_all('incm')]
    rsdArea = [item.string for item in soup.find_all('rsdArea')]
    crdtSc = [item.string for item in soup.find_all('crdtSc')]
    anin = [item.string for item in soup.find_all('anin')]
    housHoldCnt = [item.string for item in soup.find_all('housHoldCnt')]
    housAr = [item.string for item in soup.find_all('housAr')]
    lnTgtHous = [item.string for item in soup.find_all('lnTgtHous')]
    rfrcCnpl = [item.string for item in soup.find_all('rfrcCnpl')]
    grnInst = [item.string for item in soup.find_all('grnInst')]
    jnMthd = [item.string for item in soup.find_all('jnMthd')]
    rpymdCfe = [item.string for item in soup.find_all('rpymdCfe')]
    lnIcdcst = [item.string for item in soup.find_all('lnIcdcst')]
    ovItrYr = [item.string for item in soup.find_all('ovItrYr')]
    prftAddIrtCond = [item.string for item in soup.find_all('prftAddIrtCond')]
    etcRefSbjc = [item.string for item in soup.find_all('etcRefSbjc')]
    hdlInst = [item.string for item in soup.find_all('hdlInst')]
    cnpl = [item.string for item in soup.find_all('cnpl')]
    rltSite = [item.string for item in soup.find_all('rltSite')]
    tgtFltr = [item.string for item in soup.find_all('tgtFltr')]
    hdlInstDtlVw = [item.string for item in soup.find_all('hdlInstDtlVw')]
    prdExisYn = [item.string for item in soup.find_all('prdExisYn')]
    cv19Rfrc = [item.string for item in soup.find_all('cv19Rfrc')]
    prdCtg = [item.string for item in soup.find_all('prdCtg')]
    prdNm = [item.string for item in soup.find_all('prdNm')]
    cv19SuprCtg = [item.string for item in soup.find_all('cv19SuprCtg')]
    cv19SuprCtn = [item.string for item in soup.find_all('cv19SuprCtn')]
    cv19SuprTgtDtlCtn = [item.string for item in soup.find_all('cv19SuprTgtDtlCtn')]
    mgmDln = [item.string for item in soup.find_all('mgmDln')]
    prdCtg2 = [item.string for item in soup.find_all('prdCtg2')]
    fileWrtDt = [item.string for item in soup.find_all('fileWrtDt')]

    result = dict(basYm=basYm,
    snq=snq,
    finPrdNm=finPrdNm,
    lnLmt=lnLmt,
    irtCtg=irtCtg,
    irt=irt,
    maxTotLnTrm=maxTotLnTrm,
    maxDfrmTrm=maxDfrmTrm,
    maxRdptTrm=maxRdptTrm,
    rdptMthd=rdptMthd,
    usge=usge,
    trgt=trgt,
    instCtg=instCtg,
    ofrInstNm=ofrInstNm,
    rsdAreaPamtEqltIstm=rsdAreaPamtEqltIstm,
    suprTgtDtlCond=suprTgtDtlCond,
    age=age,
    incm=incm,
    rsdArea=rsdArea,
    crdtSc=crdtSc,
    anin=anin,
    housHoldCnt=housHoldCnt,
    housAr=housAr,
    lnTgtHous=lnTgtHous,
    rfrcCnpl=rfrcCnpl,
    grnInst=grnInst,
    jnMthd=jnMthd,
    rpymdCfe=rpymdCfe,
    lnIcdcst=lnIcdcst,
    ovItrYr=ovItrYr,
    prftAddIrtCond=prftAddIrtCond,
    etcRefSbjc=etcRefSbjc,
    hdlInst=hdlInst,
    cnpl=cnpl,
    rltSite=rltSite,
    tgtFltr=tgtFltr,
    hdlInstDtlVw=hdlInstDtlVw,
    prdExisYn=prdExisYn,
    cv19Rfrc=cv19Rfrc,
    prdCtg=prdCtg,
    prdNm=prdNm,
    cv19SuprCtg=cv19SuprCtg,
    cv19SuprCtn=cv19SuprCtn,
    cv19SuprTgtDtlCtn=cv19SuprTgtDtlCtn,
    mgmDln=mgmDln,
    prdCtg2=prdCtg2,
    fileWrtDt=fileWrtDt)
    
    final_result.append(result)
    
    if page < total_pages:
        page += 1
    else:
        break
```

```python
final_df = pd.DataFrame()
for item in final_result:
    temp = pd.DataFrame(item)
    final_df = pd.concat([final_df, temp])
final_df
    
```

# API 명세를 참고해서 key 리스트 만들고 딕셔너리 함수를 이용해서 만들기

```python
keys = temp.split("\n")
keys
```

```python
result_dfs = []
page = 1
while True:
    url = "http://apis.data.go.kr/1160100/service/GetSmallLoanFinanceInstituteInfoService/getOrdinaryFinanceInfo"
    payload = dict(serviceKey=os.getenv('servicekey'), pageNo=page, numOfRows=1000)
    r = requests.get(url, params=payload)
    print(r.status_code)
    data = r.text
    soup = BeautifulSoup(data, 'xml')
    total_pages = int(soup.find('totalCount').string) // int(soup.find('numOfRows').string) + 1
    print(f"{page}/{total_pages} 수집중", end="\r")
    result2 = {}
    for key in keys:
        for item in soup.find_all(key):
            result2.setdefault(key, []).append(item.string)
    result_dfs.append(pd.DataFrame(result2))

    if page < total_pages:
            page += 1
    else:
        break

final_df = pd.DataFrame()           
for df in result_dfs:
    final_df = pd.concat([final_df, df])
final_df    
```

# API응답에서 xml tag를 추출해 자동으로 데이터 정리하기

```python
result_dfs = []
page = 1
while True:
    url = "http://apis.data.go.kr/1160100/service/GetSmallLoanFinanceInstituteInfoService/getOrdinaryFinanceInfo"
    payload = dict(serviceKey=os.getenv('servicekey'), pageNo=page, numOfRows=1000)
    r = requests.get(url, params=payload)
    print(r.status_code)
    data = r.text
    soup = BeautifulSoup(data, 'xml')
    total_pages = int(soup.find('totalCount').string) // int(soup.find('numOfRows').string) + 1
    print(f"{page}/{total_pages} 수집중", end="\r")
    keys = {}
    for tags in soup.find_all()[10:]:
        if tags.name == 'item':
            continue
        else:
            keys.setdefault(f"{tags.name}", "")
    result2 = {}
    for key in keys:
        for item in soup.find_all(key):
            result2.setdefault(key, []).append(item.string)
    result_dfs.append(pd.DataFrame(result2))

    if page < total_pages:
            page += 1
    else:
        break

final_df = pd.DataFrame()           
for df in result_dfs:
    final_df = pd.concat([final_df, df])
final_df    
```
