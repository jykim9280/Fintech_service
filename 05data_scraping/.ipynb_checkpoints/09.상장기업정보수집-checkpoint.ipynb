```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
url = "https://kind.krx.co.kr/corpgeneral/corpList.do"
payload = dict(method='searchCorpList',pageIndex=1,currentPageSize=15,comAbbrv="",beginIndex="",orderMode='3',orderStat='D',isurCd="",repIsuSrtCd="",searchCodeType="",marketType="",searchType='13',industry="",fiscalYearEnd='all',comAbbrvTmp="",location='all')
r = requests.get(url, params = payload)
print(r.url)
soup = bs(r.text, 'lxml')
```

```python
print(soup)
```

```python
#for 중첩문 
company = {}
# for data in soup:
for title in soup.find_all('td', 'first'):
    company.setdefault("title", []).append(title.text.strip())
for sort in soup.find_all('td', 'textOverflow')[::2]:
    company.setdefault("sort", []).append(sort.text)
for pd in soup.find_all('td', 'textOverflow')[1::2]:
    company.setdefault("pd", []).append(pd.text)
for date in soup.find_all('td', 'txc')[0::5]:
    company.setdefault("date", []).append(date.text)
for month in soup.find_all('td', 'txc')[1::5]:
    company.setdefault("month", []).append(month.text)
for person in soup.find_all('td', 'txc')[2::5]:
    company.setdefault("person", []).append(person.text)
for link in soup.find_all('td', 'txc')[3::5]:
    company.setdefault("link", []).append(link.text)
for site in soup.find_all('td', 'txc')[4::5]:
    company.setdefault("site", []).append(site.text)
print(company)
print(type(company))
```

```python
for key, values in company.items():
    company[key] = ["-" if value == "" else value for value in values]
print(company)
```

```python
T = soup.find('table', "list type-00 tmt30")['summary']
column_names = T.split(', ')
print(len(column_names))
print(T)
```

```python
# df = pd.DataFrame(company)
# print(df)
```

```python
lengths = [len(v) for v in company.values()]
max_length = max(lengths)

for key in company:
    if len(company[key]) < max_length:
        company[key] += [np.nan] * (max_length - len(company[key]))

df = pd.DataFrame(company)
df
```

```python
keys = soup.select_one("table.list.type-00.tmt30")['summary'].split(", ")
result = {}
for tr in soup.select('tr'):
    for idx, (key, td) in enumerate(zip(keys, tr.select('td'))):
        if idx == 0:
            kinds = [img['alt'] for img in td.select('img')]
            kind = " ,".join(kinds)
            result.setdefault("증권 종류", []).append(kind)
            result.setdefault(key, []).append(td.text)
        elif idx == 6:
            td.select_one('a')['href'] if td.string == None else ""
            result.setdefault(key, []).append(td.text)
```
