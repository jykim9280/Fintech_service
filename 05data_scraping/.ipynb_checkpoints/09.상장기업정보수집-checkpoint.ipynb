```python
#for 중첩문 
company = {}
# for data in soup:
for title in soup.find_all('td', 'first'):
    company.setdefault("title", []).append(title.text.strip())
for sort in soup.find_all('td', 'textOverflow')[::2]:
    company.setdefault("sort", []).append(sort.text)
for pd in soup.find_all('td', 'textOverflow')[1::2]:
    company.setdefault("pd", []).append(pd.text)
for date in soup.find_all('td', 'txc')[0::5]:
    company.setdefault("date", []).append(date.text)
for month in soup.find_all('td', 'txc')[1::5]:
    company.setdefault("month", []).append(month.text)
for person in soup.find_all('td', 'txc')[2::5]:
    company.setdefault("person", []).append(person.text)
for link in soup.find_all('td', 'txc')[3::5]:
    company.setdefault("link", []).append(link.text)
for site in soup.find_all('td', 'txc')[4::5]:
    company.setdefault("site", []).append(site.text)
print(company)
print(type(company))
```

```python
keys = soup.select_one("table.list.type-00.tmt30")['summary'].split(", ")
```

```python
import time
```

```python
page = 1
page_size = 100
final_result_df = pd.DataFrame()
while True:
    url = "https://kind.krx.co.kr/corpgeneral/corpList.do"
    payload = dict(method='searchCorpList',pageIndex=page, currentPageSize=page_size,orderMode=3,orderStat='D',searchType=13, fiscalYearEnd='all', location='all')
    r = requests.get(url, params=payload)
    print(r.status_code)
    soup = bs(r.text, 'lxml')
    total_items = int(soup.select_one(".info.type-00 > em").text.replace(",", ""))
    total_pages = total_items // page_size + 1
    print(f"{page}/{total_pages} 수집중", end="\r")
    result = {}
    for tr in soup.select('tr'):
        for idx, (key, td) in enumerate(zip(keys, tr.select('td'))):
            if idx == 0:
                kinds = [img['alt'] for img in td.select('img')]   # 1번째 증권 종류, 회사이름
                kind = ", ".join(kinds)
                result.setdefault('증권종류', []).append(kind)
                result.setdefault(key, []).append(td.text)
            elif idx == 6:
                home_link = td.select_one('a')['href'] if td.string == None else ""  # 6번째 링크 찾기
                result.setdefault(key, []).append(home_link)
            else:
                result.setdefault(key, []).append(td.text)
    result_df = pd.DataFrame(result)
    final_result_df = pd.concat([final_result_df, result_df])
        
    if page < total_pages:
        page += 1
        time.sleep(5)
    else:
        break


final_result_df
```
