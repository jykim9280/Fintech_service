# 전국 아파트 매매 데이터 1999-2024 합치기

```python
import pandas as pd
import glob

# 각 월별 파일 경로 불러오기
all_files = []

for i in range(1, 10):
    output1 = glob.glob(f'data/*0{i}.csv')
    all_files.extend(output1)  # 모든 파일 경로를 리스트에 추가

# 파일을 읽어서 데이터프레임으로 저장 (sep=None로 구분자 자동 감지)
data_frames = [pd.read_csv(file,encoding='cp949', skiprows=15) for file in all_files]

# 하나의 데이터프레임으로 합치기
combined_df = pd.concat(data_frames, ignore_index=True)

# 데이터 확인
print(combined_df.head())
```

```python
import pandas as pd
import glob

# 각 월별 파일 경로 불러오기
all_files2 = []

for i in range(10, 13):
    output2 = glob.glob(f'data/*{i}.csv')
    all_files2.extend(output2)  # 모든 파일 경로를 리스트에 추가

# 파일을 읽어서 데이터프레임으로 저장 (sep=None로 구분자 자동 감지)
data_frames2 = [pd.read_csv(file,encoding='cp949', skiprows=15) for file in all_files2]

# 하나의 데이터프레임으로 합치기
combined_df2 = pd.concat(data_frames2, ignore_index=True)

# 데이터 확인
print(combined_df2.head())
```

```python
# '계약날짜'와 '날짜'를 datetime 형식으로 변환
apt_with_count['계약날짜'] = pd.to_datetime(apt_with_count['계약날짜'])
ex_data['날짜'] = pd.to_datetime(ex_data['날짜'])

# 토요일 또는 일요일인 데이터에서 NaN 발생 시 금요일의 데이터를 적용
def fill_weekend_with_friday(row, ex_data):
    if pd.isnull(row['USD_살때_환율']) and row['계약날짜'].weekday() in [5, 6]:  # 토요일(5) 또는 일요일(6) 이면서 NaN 값일 경우
        # 해당 계약날짜에서 가장 가까운 금요일 날짜로 데이터 가져오기
        last_friday = row['계약날짜'] - pd.DateOffset(days=(row['계약날짜'].weekday() - 4))
        friday_row = ex_data[ex_data['날짜'] == last_friday]
        if not friday_row.empty:
            return friday_row['USD_살때_환율'].values[0]
    return row['USD_살때_환율']

# 토요일 및 일요일인 경우 금요일 환율로 대체
merged_data['USD_살때_환율'] = merged_data.apply(fill_weekend_with_friday, axis=1, ex_data=ex_data)

# 최종 결과 확인
print(merged_data.head())
print(f"Number of rows in final merged data: {len(merged_data)}")
```

# 금리 정보 합치기

```python
ir_data= pd.read_csv("./data/기준금리1999-2023.csv")
```

```python
# 두 컬럼을 합쳐 yyyy-mm-dd 형식으로 변환
ir_data['DATE'] = ir_data['변경일자'].astype(str) + '-' + ir_data['Unnamed: 1'].str.replace('월 ', '-').str.replace('일', '').str.strip()

# 변경된 날짜 컬럼 확인
ir_data.head(55)
ir_data = ir_data.drop(['변경일자','Unnamed: 1'], axis=1)
```

```python
ir_data
```

```python
apt_academy_exchabge_merged_data
```

```python
# 계약날짜와 DATE를 datetime 형식으로 변환
apt_academy_exchabge_merged_data['계약날짜'] = pd.to_datetime(apt_academy_exchabge_merged_data['계약날짜'])
ir_data['DATE'] = pd.to_datetime(ir_data['DATE'])

# 계약날짜 기준으로 backward merge를 수행하여 ir_data의 기준금리를 apt_academy_exchabge_merged_data에 추가
final_merged_data = pd.merge_asof(
    apt_academy_exchabge_merged_data.sort_values('계약날짜'),
    ir_data[['DATE', '기준금리']].sort_values('DATE'),
    left_on='계약날짜',
    right_on='DATE',
    direction='backward'
)

# 필요한 경우 DATE 컬럼을 제거 (이미 병합되었으므로 필요 없을 경우)
final_merged_data = final_merged_data.drop(columns=['DATE'])

# 결과 확인
print(final_merged_data.head())
```
