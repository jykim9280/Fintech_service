```python
data.columns = [col.replace(" ","").replace(".", "") for col in data.columns] # 컬럼명 여백 제거
data.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop('YearlyAmountSpent', axis = 1)
y = data['YearlyAmountSpent']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=7)
```

```python
# 다중선형회귀분석하기
```

```python
from statsmodels.formula.api import ols
```

```python
train_data = pd.concat([X_train, y_train], axis = 1)
train_data.head()
```

```python
train_data.columns
```

```python
lm = ols('YearlyAmountSpent ~ AvgSessionLength + TimeonApp + TimeonWebsite + LengthofMembership + state_name', train_data).fit() 
#R은 종속변수먼저, 이후 독립변수 나열해야 함
```

```python
lm.summary()
```

```python
# 수정된 결정계수는 0.984 (독립변수가 많을수록 회귀계수는 자동으로 올라가므로 '엄청 좋은 모델이네!'라고 생각하면 안 됨)
# Prob (F-statistic), 즉 p-value 4.36e-279로 통계적으로 유의미한 모델임
# 사용한 독립변수에서 p-value 0.05 이하인 값들만 솎아줌
# AR 0.049     11.2649
# MI 0.030     9.6009
# MT 0.013     12.1776
# AvgSessionLength, TimeonWebsite, LengthofMembership 만으로 모델 돌려도 충분할 것으로 추측됨
# Kurtosis:	2.946 만약 4가 나오면 회귀분석이 아니고 시계열 분석이 되어야 함
```

```python
pred = lm.predict(X_test)
```

```python
pred
```

```python
from sklearn.metrics import mean_squared_error #평균제곱오차
```

```python
print("Mean_Squared_Error(MSE): ", mean_squared_error(y_test, pred))
print("Root_Mean_Squared_Error(RMSE): ", np.sqrt(mean_squared_error(y_test, pred))) 
```

```python
# p-value 가져오기
p_value = lm.pvalues[1]
```

```python
# 산점도와 회귀선 그리기
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_test, y=pred, label='data_points')
sns.regplot(x=y_test, y=pred, scatter=False, label="Regression Line", color='red')
plt.xlabel("Actual Yearly Amount Spent")
plt.ylabel("pred Yearly Amount Spent")
plt.legend()
plt.show()
```

```python
X = pd.get_dummies(X, drop_first=True)
X
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler()
X_scaled = mms.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled
```

```python
# statsmodels의 python문법
import statsmodels.api as sm
```

```python
X2_train, X2_test, y2_train, y2_test = train_test_split(X_scaled, y, test_size=0.2, random_state=7)
```

```python
X2_train
```

```python
# 상수항 추가
X2_train = sm.add_constant(X2_train)
X2_test = sm.add_constant(X2_test)
```

```python
X2_train.head(2)
```

```python
X2_train.shape
```

```python
X2_test.shape
```

```python
lr = sm.OLS(y2_train, X2_train).fit()
```

```python
lr.summary()
```

```python
pred2 = lr.predict(X2_test)
```

```python
print("MSE:", mean_squared_error(y2_test, pred2))
print("RMSE:", np.sqrt(mean_squared_error(y2_test, pred2)))
```

```python
lr.pvalues
```

```python
p_value2 = lr.pvalues[1]
```

```python
# 산점도와 회귀선 그리기
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y2_test, y=pred2, label='data_points')
sns.regplot(x=y2_test, y=pred2, scatter=False, label="Regression Line", color='red')
plt.xlabel("Actual Yearly Amount Spent")
plt.ylabel("pred Yearly Amount Spent")
plt.legend()
plt.show()
```
