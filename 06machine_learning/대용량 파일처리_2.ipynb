```python
import pandas as pd
import glob

# 각 월별 파일 경로 불러오기
all_files = []

for i in range(1, 10):
    output1 = glob.glob(f'data/*0{i}.csv')
    all_files.extend(output1)  # 모든 파일 경로를 리스트에 추가

# 파일을 읽어서 데이터프레임으로 저장 (sep=None로 구분자 자동 감지)
data_frames = [pd.read_csv(file,encoding='cp949', skiprows=15) for file in all_files]

# 하나의 데이터프레임으로 합치기
combined_df = pd.concat(data_frames, ignore_index=True)

# 데이터 확인
print(combined_df.head())
```

```python
# 각 월별 파일 경로 불러오기
all_files2 = []

for i in range(10, 13):
    output2 = glob.glob(f'data/*{i}.csv')
    all_files2.extend(output2)  # 모든 파일 경로를 리스트에 추가

# 파일을 읽어서 데이터프레임으로 저장 (sep=None로 구분자 자동 감지)
data_frames2 = [pd.read_csv(file,encoding='cp949', skiprows=15) for file in all_files2]

# 하나의 데이터프레임으로 합치기
combined_df2 = pd.concat(data_frames2, ignore_index=True)

# 데이터 확인
print(combined_df2.head())
```

```python
import pandas as pd
import time

# 시간 기록 시작
start_time = time.time()

# 진행 상황 확인을 위한 메시지 출력
print("Concatenation started...")

# 데이터프레임 결합
apt_result = pd.concat([combined_df, combined_df2], axis=0)

# 시간 기록 종료
end_time = time.time()

# 완료 메시지 및 소요 시간 출력
print(f"Concatenation completed in {end_time - start_time:.2f} seconds.")
```

```python
# '시도교육청명'과 '행정구역명'이 동일한 것들의 개수를 구함
academy_count = academy_result.groupby(['시도교육청명', '행정구역명']).size().reset_index(name='count')

# 결과 확인
print(academy_count.head(30))
```

```python
# 두 데이터프레임 병합 (시도교육청명과 시군구 기준으로 count 값 붙이기)
apt_result_with_count = pd.merge(
    apt_result, 
    academy_count[['시도교육청명', '행정구역명', 'count']], 
    left_on=['시도교육청명', '시군구'], 
    right_on=['시도교육청명', '행정구역명'], 
    how='left'
)
```

```python
apt_result_with_count = apt_result_with_count[['시도교육청명','시군구','단지명','거래금액(만원)','전용면적(㎡)','층','건축년도','계약년월','계약일','count']]
```
