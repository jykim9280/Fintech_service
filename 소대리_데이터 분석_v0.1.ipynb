# 한국 표준 산업 체계

```python
# 엑셀 파일 불러오기
file_path = './data/한국표준산업분류체계.xlsx'
classification_data = pd.read_excel(file_path)

# 필요한 열만 추출 (코드1, 코드5, 항목명5)
selected_columns = classification_data[['코드1', '코드5', '항목명5']].dropna()

# 결과 확인
print(selected_columns.head(10))
```

```python
import pandas as pd

# '코드1'과 '코드5'를 키로, '항목명5'를 값으로 하는 딕셔너리 생성
classification_data['키'] = classification_data['코드1'].astype(str) + classification_data['코드5'].astype(str)
code_to_name = dict(zip(classification_data['키'], classification_data['항목명5']))

# insolvency_rate_data에서도 '업종코드'와 '코드5'를 연결하여 키 생성
insolvency_rate_data['키'] = insolvency_rate_data['업종코드']

# '키' 값을 이용해 '항목명5'로 대체
insolvency_rate_data['업종코드'] = insolvency_rate_data['키'].map(code_to_name)

# 결과 확인
print(insolvency_rate_data[['업종코드']].head(10))
```

```python
insolvency_rate_new_data = insolvency_rate_data.copy()
insolvency_rate_new_data.head(10)
```
