### 매번 검색어 마다 total 개수 확인할 수 없으니 몇 번 반복할지 모를 때, while문 사용

```python
import time
```

```python
book_list2=[]
page = 1

while True:
    client_id = "VfCSVYpBxY6EEAWW1zcr" #네이버 api에 접속 가능한 id
    client_secret = "bTXn2nJ3Ig" 
    encText = urllib.parse.quote("자취생")
    print(encText)
    url = f"https://openapi.naver.com/v1/search/book.json?query={encText}&display=10&start={10*page-9}"
    print("url:", url)
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    if(rescode==200):
        response_body = response.read()
        data = json.loads(response_body.decode('utf-8'))
        book_list2.append(data)
        total_pages = data['total'] // 10 + 1
        page += 1
    else:
        print("Error Code:" + rescode)
        break
    if page > total_pages: #페이지 여분 마지막 자투리까지 가져오기 위해서
        print("완료")
        break
print(len(book_list2))
```

```python
#강사님 코드
book_lists = []
page = 1
start = 1
query = input("검색어를 입력하세요: ")
while True:
    client_id = "pRmtpgiw8Lf7F3QGZ4ar" # 네이버 api에 접속 가능한 id 
    client_secret = "M_XHcT1MXF"       # 네이버 api에 접속 가능한 pw 
    encText = urllib.parse.quote(query)
#     print(encText)
    url = f"https://openapi.naver.com/v1/search/book.json?query={encText}&display=10&start={start}"
#     print("url:", url, end="\r")
    try:
        request = urllib.request.Request(url)
        request.add_header("X-Naver-Client-Id",client_id)
        request.add_header("X-Naver-Client-Secret",client_secret)
        response = urllib.request.urlopen(request)
        rescode = response.getcode()
        if(rescode==200):
            response_body = response.read()
            data = json.loads(response_body.decode('utf-8'))
            book_lists.append(json.loads(response_body.decode('utf-8')))
            total_page = data['total'] // 10 + 1

        else:
            print("Error Code:" + rescode)
            break

        if page < total_page:
            page += 1
            if start != 991:
                start += 10
            elif start == 991:
                start += 9
            print(f"{page:03d}/{total_page:03d}, start: {start} 추출중", end="\r")
        else:
            break
        time.sleep(0.5)
    except Exception as e:
        print(e)
        break
    
    
print(len(book_lists))
result = pd.DataFrame()
for book_list in book_lists:
    temp = pd.json_normalize(book_list['items'])
    result = pd.concat([result, temp])
result
result.to_csv(f"naver_book_api_fintech_{query}_result.csv", encoding="utf-8")
```

```python
# 1. json으로 온 자료를 엑셀 형태로 정리하기 dataframe
# 2. parameter 추가로 입력해서 전체 검색결과 모아서 합치기

# 1. 서비스 선택 : 블로그 책 뉴스 전문자료
# 2. 검색어 입력
# 3. 함수화
```

```python
def search():
    book_lists = []
    page = 1
    start = 1
    service = input("카테고리를 입력하세요(blog,book,news,doc): ").lower()
    query = input("검색어를 입력하세요: ")
    while True:
        client_id = "pRmtpgiw8Lf7F3QGZ4ar" # 네이버 api에 접속 가능한 id 
        client_secret = "M_XHcT1MXF"       # 네이버 api에 접속 가능한 pw 
        encText = urllib.parse.quote(query)
        url = f"https://openapi.naver.com/v1/search/{service}.json?query={encText}&display=10&start={start}"
        try:
            request = urllib.request.Request(url)
            request.add_header("X-Naver-Client-Id",client_id)
            request.add_header("X-Naver-Client-Secret",client_secret)
            response = urllib.request.urlopen(request)
            rescode = response.getcode()
            if(rescode==200):
                response_body = response.read()
                data = json.loads(response_body.decode('utf-8'))
                book_lists.append(json.loads(response_body.decode('utf-8')))
                total_page = data['total'] // 10 + 1
            else:
                print("Error Code:" + rescode)
                break
                
            if page < total_page:
                page += 1
                if start != 991:
                    start += 10
                elif start == 991:
                    start += 9
                print(f"{page:03d}/{total_page:03d}, start: {start} 추출중", end="\r")
            else:
                break
            time.sleep(0.5)
            
        except Exception as e:
            print(e)
            break    
    result = pd.DataFrame()
    for book_list in book_lists:
        temp = pd.json_normalize(book_list['items'])
        result = pd.concat([result, temp])
    print(result)

search()  
# print(len(book_lists))
# result = pd.DataFrame()
# for book_list in book_lists:
#     temp = pd.json_normalize(book_list['items'])
#     result = pd.concat([result, temp])
# result
# result.to_csv(f"naver_news_api_fintech_{query}_result.csv", encoding="utf-8")
```
