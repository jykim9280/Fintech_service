# 반복문 for, while
* 반복문은 동일한 작업을 여러 번 반복해야 할 경우 사용하는 명령
* for, while 두 종류의 반복문이 있음.
* for는 반복 횟수가 정해져 있는 경우에 사용.
* while은 반복 횟수가 정해지지 않은 경우에 사용.

## for 반복문 : 반복할 횟수가 정해져 있는 경우
for 변수 in 반복할 객체, 자료:<br>
____반복할 코드

```python
# for 문을 이용해서 1-10 숫자 출력
for i in range(1,11):
    print(i, end =" ")
```

* 순서가 있는 자료형은 모두 for 문에 사용 가능
    * 문자열, list, tuple

```python
for char in "파이썬":
    print(char, end=" ")
```

```python
for item in [1, 2, 3, 4, 5]:
    print(item, end = " ")
```

```python
L = [1,2,3,3, "문자열", [1,2,3], (4,5,6), {'key':'value'}]
```

```python
for item in L:
    print(item, end = " ")
```

```python
T = tuple(L)
for item in T:
    print(item, end = " ")
```

```python
d = dict(번호=1, 주소='서울', 전화번호='010-1234-5678', 이름='홍길동')
for key in d:
    print(d[key], end=" ")
#딕셔너리는 반복문에 넣었을 때 key만 출력되므로, value 넣기 위해서는 key로 value 호출하기
```

```python
for key in d:
    print(d.get(key,0), end = " ")
#key가 없으면 0으로 출력해라.
```

```python
for k,v in d.items():
    print(k,v)
```

# while 반복할 횟수가 정해지지 않은 경우
* 무한 반복 가능
* 무한 반복 되지 않도록 조건식을 잘 짜주는 것이 중요

```python
#while문을 이용해서 1-10까지 더해서 결과를 출력하세요.
i = 1
while i <= 10:
    print(i, end=" ")
    i += 1
```

```python
# 웹 크롤링시, while 많이 사용
# while문을 이용해서 1-10까지 더해서 결과를 출력하세요.
i = 1
hap = 0
while i <= 10:
    hap += i
    i += 1
print(hap)
```

```python
#20에서 1까지 숫자를 더한 값을 출력하세요.
i = 20
hap = 0
while i > 0:
    hap += i
    i -= 1
print(hap)
```

```python
#1-50 까지 수 중에서 3의 배수이거나, 짝수인 수를 찾아서 더한 값을 출력하시오.
i = 1
hap = 0
while i <= 50:
    if i % 3 == 0 or i % 2 ==0:
        hap += i
    i += 1
print(hap)
```

반복문에서 아무일도 하지 않고 건너뛰기 : pass
* 우선 반복할 코드가 들어가야 하는 곳을 만들고 나중에 작업하기 위해 사용

```python
i = 1
while i <= 10:
    pass
    i += 1
```

Break 반복문을 정지시키는 명령

```python
for i in range(1.101):
    if i > 50:
        break
    else:
        print(i, end=" ")
```

# continue: 특정 조건에서 코드 실행 건너뛰기

```python
#1~100까지 수 중에서 짝수면 더하고 홀수면 건너뛰기
hap = 0
for i in range(1, 101):
    if i % 2 == 1:
        continue
    else:
        hap += i
        print(hap, end =" ")
```

```python
#1~100까지 수 중에서 짝수면 더하고 홀수면 건너뛰기
i = 1
hap = 0

while i <= 100: 
    if i % 2 == 1:
        i += 1
        continue
    hap += i
    i += 1
    print(hap, end =" ")
```

```python
total = 0
i = 1

while True:
    if i > 100:
        break
    if i % 2 == 0:
        total += i
        print(total, end =" ")
    i += 1
```

```python
## 중첩 반복문
* 반복문 2개 이상을 겹쳐서 사용
* 바늘 시계의 원리와 같음. (분침이 0~59까지 돌아야 시침이 한 칸 움직임)
* 안쪽 반복이 끝난 후 바깥쪽 반복문이 실행
```

```python
for si in range(0, 24):
    for bun in range(0, 60):
        print(si, "시", bun, "분", end= " ")
```

```python
for i in range(2,10):
    for j in range(1,10):
        print(i,'x',j,'=', i*j)
```
