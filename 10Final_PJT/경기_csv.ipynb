```python
import pandas as pd
data = pd.read_csv("../data/소상공인시장진흥공단_상가(상권)정보_경기_202409.csv")
```

```python
data.info()
```

```python
# 조건에 따라 데이터 필터링
filtered_data = data[(data["상권업종대분류코드"] == "I2") & (data["시도코드"].astype(str).str.startswith("41"))]

# 필터링된 데이터 출력
print(filtered_data)
```

```python
# DataFrame을 JSON 파일로 변환하고 저장
json_file_path = "csv_data.json"
filtered_data.to_json(json_file_path, orient="records", force_ascii=False, indent=4)

print(f"데이터가 JSON 파일로 '{json_file_path}'에 저장되었습니다.")
```
