# 간단한 시계열 데이터 만들기
- 정상 시계열(우연 변동 시계열)

```python
dd = np.array([1342, 1442, 1252, 1343, 1425, 1362, 1256, 1272, 1234, 1359, 1412, 1253, 1201, 1478, 1322, 1406, 1254, 1289, 1497, 1208]).reshape(5,4)
dd
```

```python
dd_ts = pd.DataFrame(dd, index=range(2016,2021), columns=['Q1', 'Q2', 'Q3', 'Q4'])
```

```python
dd_ts
```

데이터프레임 dd_ts의 행과 열을 stack(스택)방식으로 변환하고 인덱스 초기화

```python
dd_ts.stack()
```

```python
dd_ts = dd_ts.stack().reset_index()
dd_ts
```

```python
dd_ts.colums=['year', 'quater', 'value']
dd_ts
```
