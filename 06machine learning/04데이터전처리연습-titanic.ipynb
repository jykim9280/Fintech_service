### 2) 평균 대치법 : 결측치가 있는 컬럼에서 데이터의 평균을 구한 후 결측값을 대치
* 평균을 이용하기 때문에 간편
* 데이터에 이상치가 있을 경우 평균을 이용할 수 없어서, 중앙값이나 최빈값을 고려해야 함

```python
#평균
a = np.array([24, 5, 10, 34, 20, 18, 28, 20])
a.mean()
```

```python
#평균
b = np.array([24, 5, 10, 34, 20, 18, 28, 200]) #a와 달리이상치 200 삽입
b.mean()
```

```python
#중앙값
c = np.array([24, 5, 10, 34, 20, 18, 28, 20])
np.median(c)
```

```python
#중앙값
d = np.array([24, 5, 10, 34, 20, 18, 28, 200])
np.median(d)
```

전체 데이터 프레임에서 Age 컬럼에 결측값이 있는 행만 추출

```python
data[data['Age'].isna() == True]
```

결측 값이 있는 컬럼을 나중에 다시 확인하고 싶을 때<br>
1. isna() == True로 결측값이 있는 컬럼을 추출한다.<br>
2. 추출된 데이터 프레임의 index 추출 .index
3. 추출한 인덱스를 변수에 저장
4. 데이터프레임.iloc[index]로 조회

```python
na_index = data[data['Age'].isna() == True].index #추출한 인덱스를 변수에 저장
```

```python
data.iloc[na_index]
```

* 첫번째, .fillna(대치할 값) 로 평균 대치 하기

```python
data['Age'].fillna(round(data['Age'].mean()))
```

```python
data['Age'].fillna(round(data['Age'].median()))
```

```python
#최빈값으로 대치하기 mode
data['Age'].mode()
```

```python
data['Age'].fillna(data['Age'].mode())
```

* 두번째, scikit-learn 라이브러리의 SimpleImputerr: 평균, 중앙값, 최빈값 대치

```python
from sklearn.impute import SimpleImputer
```

```python
#평균으로 대치 mean
imp_mean = SimpleImputer(strategy = 'mean')
imp_mean.fit_transform(data['Age'].values.reshape(-1,1))[:,0]
```

```python
#중앙값으로 대치 median
imp_mean = SimpleImputer(strategy = 'median')
imp_mean.fit_transform(data['Age'].values.reshape(-1,1))[:,0]
```

* .fillna(대치할 값)으로 평균 대치하기

```python
data_mean = data.copy()
data_median = data.copy()
data_mode = data.copy()
```

```python
#mean 30
data_mean['Age'] = data['Age'].fillna(round(data['Age'].mean()))
```

```python
data_mean['Age']
```

```python
data_mode.iloc[na_index]
```

```python
* KNN(K-NearestNeighbors) 최근접이웃 대치법
```
